```python
dates = [x.text for (i, x) in enumerate(wapo.find_all('span', {'class': 'flex nowrap font-light gray-dark italic font-xxxxs'})) if i % 2 == 0]
headlines = [x.text for x in wapo.find_all('a', {'class': 'font-md font--magazine-headline font-bold mb-sm black'})]
desc = [x.text for x in wapo.find_all("div", {"class" :"font-body font-light font-xxs antialiased gray-dark lh-1 mt-sm mb-sm"})]
news = pd.DataFrame(columns=["date", "headline", "desc"])
news['date'] = dates
news['headline'] = headlines
news['desc'] = desc
news.date = pd.to_datetime(news.date, format='mixed')
news.to_csv(news_csv_path, index=False)
```
